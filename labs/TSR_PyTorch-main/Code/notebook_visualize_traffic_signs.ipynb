```python
"""
This notebook is to display the sample image of all classes of the GTSRB dataset
"""
```

```python
import os
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# 

numClasses = 43

fig, axs = plt.subplots(9,5, figsize=(100,150))
fig.tight_layout(h_pad = 20)
for i in range(numClasses):
    row = i // 5
    col = i % 5
    
    imgName = '../GTSRB/Meta/' + str(i) + ".png"
    img = Image.open(imgName)
    axs[row, col].imshow(img)
    axs[row, col].set_title(str(i), fontsize=75)

plt.savefig("classes.png", bbox_inches = 'tight', pad_inches=0.5)
```
